<h1>Step_5</h1>
<h2>Data Normalization</h2>

```python
tita1.columns
```

```python
tita1.pclass = tita1.pclass.apply(lambda x : (x - tita1.pclass.min()) / tita1.pclass.max())

tita1.sex = tita1.sex.apply(lambda x : (x - tita1.sex.min()) / tita1.sex.max())

tita1.age = tita1.age.apply(lambda x : (x - tita1.age.min()) / tita1.age.max())

tita1.sibsp = tita1.sibsp.apply(lambda x : (x - tita1.sibsp.min()) / tita1.sibsp.max())

tita1.parch = tita1.parch.apply(lambda x : (x - tita1.parch.min()) / tita1.parch.max())

tita1.fare = tita1.fare.apply(lambda x : (x - tita1.fare.min()) / tita1.fare.max())

tita1.embarked = tita1.embarked.apply(lambda x : (x - tita1.embarked.min()) / tita1.embarked.max())

tita1.boat = tita1.boat.apply(lambda x : (x - tita1.boat.min()) / tita1.boat.max())
```

<h1>Step_6</h1>
<h2>Modeling</h2>

<h2>KNN Algorithm</h2>

```python
x = tita1[['boat', 'fare', 'parch']]
y = tita1['survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 123456, stratify = y)
k = 3
knn_model = KNeighborsClassifier(metric='minkowski', n_neighbors = k)
knn_model.fit(x_train, y_train)
accuracy_model = knn_model.score(x_test, y_test) * 100
print("Accuracy in this model(knn): %.3f" % accuracy_model)
```

<h1>Step_7</h1>
<h2>Model optimizing</h2>

```python
k_score = []
for k in range(3, 100, 2):
    knn_model_op = KNeighborsClassifier(n_neighbors=k, metric = 'minkowski')
    scores = cross_val_score(knn_model_op, x, y, cv =10, scoring='accuracy')
    k_score.append([k, scores.mean()])

k_score_max = max(k_score[i][1] for i in range(len(k_score)))
for i in range(len(k_score)):
    if k_score[i][1] == k_score_max:
        k_max = k_score[i][0]
        break
print("max_accuracy: %.2f" % (k_score_max * 100))
print("k_good:", k_max) 
    
```

```python
plt.figure(figsize=(20, 5))
plt.plot([k_score[i][0] for i in range(len(k_score))], [k_score[i][1] for i in range(len(k_score))])
plt.ylabel('scores')
plt.xlabel('k')
plt.xticks(range(1, 100, 2))
plt.grid()
plt.show()
```
